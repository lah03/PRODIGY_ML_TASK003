```python
import pandas as pd

train = pd.read_csv("train.csv")
train.columns = train.columns.str.strip().str.lower()

label_col = train.columns[0]

X = train.drop(label_col, axis=1)
y = train[label_col]

print(X.shape, y.shape)
```

```python
train[label_col].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Class Distribution Pie Chart")
plt.ylabel("")
plt.show()
```

```python
X.iloc[:, :50].hist(figsize=(10,6))
plt.suptitle("Pixel Value Distribution (First 50 Features)")
plt.show()
```

```python
plt.boxplot(X.iloc[:, :20])
plt.title("Boxplot of Pixel Features")
plt.show()
```

```python
import seaborn as sns

corr = X.iloc[:, :50].corr()
sns.heatmap(corr)
plt.title("Feature Correlation Heatmap")
plt.show()
```
